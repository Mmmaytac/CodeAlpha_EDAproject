```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\hp\Downloads\datasets\amazon.csv")
df
```

## Data Cleaning

```python
df.columns
```

```python
df.drop(columns=['about_product','review_id', 'review_title','review_content', 'img_link', 'product_link'], inplace=True)
```

```python
columns=[]
for i in df.columns:
    columns.append(i.capitalize())
df.columns=columns
df.columns
```

```python
df.info()
```

```python
df.Rating_count.isnull().sum()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.info()
```

## Data Normalization

```python
df.Category
```

```python
df['Main_category'] = df['Category'].str.split('|').str[0]
df.drop(columns='Category',inplace=True)
```

```python
df.Discounted_price=df.Discounted_price.apply(lambda x: x[1:])
df['Discounted_price'] = pd.to_numeric(df['Discounted_price'].str.replace(',', ''), errors='coerce')
df['Discounted_price']
```

```python
df.Actual_price=df.Actual_price.apply(lambda x: x[1:])
df['Actual_price'] = pd.to_numeric(df['Actual_price'].str.replace(',', ''), errors='coerce')
df['Actual_price']
```

```python
df.Discount_percentage=df.Discount_percentage.apply(lambda x: x[:-1])
df.Discount_percentage=df.Discount_percentage.astype('int')
```

```python
df["Rating"] = pd.to_numeric(df["Rating"], errors='coerce')
```

```python
df.Rating_count=df.Rating_count.str.replace(',','').astype('int')
```

```python
df['Discounted_price(USD)']=df.Discounted_price*0.012
df['Actual_price(USD)']=df.Actual_price*0.012
```

```python
df.drop(columns=['Discounted_price','Actual_price'],inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

# Data Visualization

## Top Performing Categories

```python
plt.figure(figsize=(8, 4))
category_sales = df['Main_category'].value_counts().head(10)
sns.barplot(x=category_sales.values, y=category_sales.index)
plt.title('Top 10 Categories by Performing')
plt.xlabel('Sales count')
plt.tight_layout()
plt.show()
```

## Price Distribution

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Actual_price(USD)'], bins=50, kde=True)
plt.title('Price Distrubition')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## Rating Distrubition

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

## Average Price by Category

```python
plt.figure(figsize=(8, 4))
category_price = df.groupby('Main_category')['Actual_price(USD)'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=category_price.values, y=category_price.index)
plt.title('Average Price by Category (Top 10)')
plt.xlabel('Avg Price')
plt.tight_layout()
plt.show()
```

## Correlation between Rating and Price

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='Rating', y='Actual_price(USD)', alpha=0.6)
plt.title('Correlation between Rating and Price')
plt.xlabel('Rating')
plt.ylabel('Price')
plt.show()
```

## Correlation between Rating, Price, and Discount Percentage

```python
df[['Discount_percentage','Rating','Actual_price(USD)']].corr()
```
